```python
CATEGORY ={
    'coughing': 'coughing',
    'silence': 'silence',
    'sneezing': 'sneezing',
    'speech': 'speech'
}
```

```python
files = list(utils.scan_files(
        root=data_root,
        recursive=True,
        pattern='*.wav',
    ))
```

```python
# Adjusted get_category() to grab the directory-part of the path, which describes the category of the file
def get_category(filename):
    category_id = filename.split('/')[0]
    return CATEGORY[category_id]
```
